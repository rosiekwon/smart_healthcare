```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#누락데이터가 500개 이상일때 열 삭제
df.dropna(axis=1,thresh=500)
```

```python
#how = 'any' ; NAN값이 하나라도 존재하면 삭제하는 명령어
df.dropna(subset=['age'],how='any',axis=0)
```

```python
most_freq = df['embark_town'].value_counts(dropna=True).idxmax() #최빈값
most_freq
```
